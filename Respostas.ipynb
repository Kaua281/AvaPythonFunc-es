1. Elaborar um programa em Python para gerar o gráfico de 10 funções de 1° e 10 funções de 2° grau, de exercícios que você já estudou, utilizando a biblioteca matplotlib, dados os seus coeficientes, nas formas:

  

 $$(a) \quad f(x) = ax + b$$

 $$(b) \quad f(x) = ax^2 + bx + c$$

```python
from pylab import arange
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.pyplot as plt

def extrair_coeficientes(funcao):
    termos = funcao.replace("-", "- ").split()  # Adiciona um espaço após cada sinal de subtração para simplificar a divisão

    # Inicializa os coeficientes
    a = b = c = 0.0

    for i, termo in enumerate(termos):
        if "x^2" in termo:
            a = float(termo.replace("x^2", "")) if termo.replace("x^2", "") else 1.0
        elif "x" in termo:
            b = float(termo.replace("x", "")) if termo.replace("x", "") else 1.0
        elif termo.replace(".", "").isdigit():
            if i > 0 and termos[i - 1] == "-":
                c = -float(termo)
            else:
                c = float(termo)

    tipo = "segundo_grau" if a != 0 else "primeiro_grau"

    return tipo, a, b, c

def raizes(tipo, coeficientes):
    valores_de_x = []
    valores_de_y = []   
    for x in range(-10, 11, 1):
        if tipo == "segundo_grau":
            a = coeficientes[0]
            b = coeficientes[1]
            c = coeficientes[2]
            valores_de_x.append(x)
            valores_de_y.append(a*(x**2) + (b*x) + c)
            
        elif tipo == "primeiro_grau":
            a = coeficientes[0]
            b = coeficientes[1]
            valores_de_x.append(x)
            valores_de_y.append((a*x) + b)
            
        else:
            print("Não foi possível extrair os coeficientes da função. Certifique-se de que a entrada está no formato correto.")
    
    return valores_de_x, valores_de_y

def grafico(valores_de_x, valores_de_y, funcao):
    with plt.style.context('dark_background'):
        fig, ax = plt.subplots()
        plt.title(funcao, fontsize=15)
        plt.xlabel("Valores de X", fontsize=12)
        plt.ylabel("Valores de Y", fontsize=12)
        ax.plot(valores_de_x, valores_de_y, linewidth=2)
        plt.scatter(valores_de_x, valores_de_y, color = 'white', marker="o")
        ax.yaxis.grid(color='gray', linestyle='-')
        plt.xticks(range(-10, 11, 1))
        plt.ylim(min(valores_de_y) - 10, max(valores_de_y) + 10)
    plt.show()

def teste_codigo():
    funcoes = [
        "2x^2 + 3x - 5",
        "x^2 + 2x + 1",
        "3x + 4",
        "x - 3",
        "-4x + 6",
        "3x - 2",
        "-x + 4",
    ]

    for funcao in funcoes:
        print(f"Função: {funcao}")
        tipo, *coeficientes = extrair_coeficientes(funcao)
        valores_de_x, valores_de_y = raizes(tipo, coeficientes)
        grafico(valores_de_x, valores_de_y, funcao)  # Adicionado o argumento 'funcao' aqui
        

# Execute a função de teste
teste_codigo()

```
